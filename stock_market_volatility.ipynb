# The Impact of Changing Social, Political, and Economic Climates on Stock Market Volatility

## Abstract

This research investigates the volatility of the stock market from 1990 to 2022, employing various regression models to analyze the impact of socioeconomic factors. Utilizing a comprehensive dataset, including a newspaper-based Equity Market Volatility (EMV) tracker, the study covers general economic and policy-related categories. The models, including Ordinary Least Squares (OLS), Lasso, Ridge, and Elastic Net regressions, reveal consistent significance of factors like inflation, interest rate, trade, business investment, sentiment, regulation, and labor disputes. Examining distinct time segments unveils nuanced patterns, aligning significant predictors with historical events. Notably, the research underscores the profound influence of social and political dynamics alongside economic factors on stock market volatility. Despite model variations, certain variables like trade and interest rate consistently impact market conditions. Noteworthy is the absence of intellectual property during the dot-com era and real estate during the housing crisis, highlighting the nuanced evolution of market dynamics over time.

## Introduction

Understanding the volatility of the stock market is crucial for making sound financial decisions, as it highlights market inefficiencies and impacts the risk-return trade-off of stocks and bonds. Such decisions are pivotal to ensure that scarce resources are allocated to the most profitable activities, and the health of the stock market can directly influence a nation’s economy.

Significant research has been conducted on building statistical models to forecast the stock market by considering numerous factors. The goal of this paper is to build on previous research and discover which socioeconomic factors impact the volatility of the stock market. It will do so by examining differences in stock market behavior across varying periods over the past few decades to study the degree to which changes in the political, economic, and social climates impacted volatility.

The data used in this paper has been gathered by Scott R. Baker, Nick Bloom, Steven J. Davis, and Kyle Kost. In this dataset, a newspaper-based Equity Market Volatility (EMV) tracker was created that moves with the Chicago Board Options Exchange’s (CBOE) VIX index, an estimator of the equity market's implied volatility. The EMV tracker has two broad categories: general economic categories and policy-related categories. The general economic categories include those reflecting the state of the labor market and entrepreneurship, such as macroeconomic news and outlook, healthcare matters, labor disputes, and intellectual property matters. Policy-related categories include matters that can be controlled by the government to impact the state of the economy, such as fiscal, monetary, and trade policies, as well as competition, labor, housing, and land management policies. The time period ranges from 1990 to 2022. The categories cover a wide range of socioeconomic factors, providing the opportunity to conduct in-depth research regarding which of these factors impact market volatility and how that changes over time.

An OLS model, LASSO regression model, ridge-regression model, and an elastic-net model were trained on the data, and the fitted models were analyzed to see which factors had the most impact in predicting volatility. Additionally, the data was split into the following six segments to be analyzed separately: January 1990 to January 1998, a period of growth that saw the rise of the dot-com bubble; July 1998 to March 2003, a recession period after the dot-com crash; April 2003 to December 2007, a period of growth that saw numerous companies emerging; January 2008 to September 2009, a period of recession caused by the housing crisis; October 2009 to December 2019, a period that saw economic growth but also political instability and rising human-rights movements; and January 2020 to December 2022, a period of recession caused by COVID-19. The goal is to determine whether the factors that impact the volatility of the market change or remain the same across these different time segments.

## Model Specification

To create a model considering all the data from 1990 to 2022, four different models were built: an Ordinary Least Squares (OLS) regression model, a regularized regression model using the Lasso method, a regularized regression model using the Ridge method, and a final regularized regularization method that combines the latter two using Elastic Net. The Root Mean Squared Error (RMSE) and R-squared were calculated to evaluate the models.

The OLS regression model was fitted first. While this model does not perform automatic feature selection, it is easily interpretable which aids in exploring which predictors impact stock market volatility. However, the OLS regression model is prone to overfitting when dealing with a large number of predictors, especially if there is collinearity among them as the ones in this dataset do. It also assumes homoscedasticity, independence of errors, and normality of residuals all of which, if violated, reduce the reliability of the model's estimates. These will be checked in the 'Training and Diagnostics' section below. 

The Ridge regression model was then built as it offers less variance than the OLS model due to the small amount of bias created by the penalty. The issue with the OLS model is that a small change in any predictor value can lead to a large change in the VIX value, reducing the interpretability of the model regarding how much a predictor impacts VIX. Cross-validation was used to select the optimal lambda values and carry out automatic feature selection.

The Lasso regression model was also built as it offers less variance than the OLS model but can exclude variables completely during feature selection. This provides a better understanding of which predictors do not impact VIX. Cross-validation was used to select the optimal lambda values here as well. This model offers a sparse matrix as seen by predictors having coefficients that are 0. Sparsity allows a model to be more interpretable because it then highlights influential features. It reduces model complexity and improves generalization performance. This is important for application in high-dimensional data. Therefore, the Lasso model is able to reduce the overfitting that exists in the OLS regression model, but has more interpretability than the Ridge regression model. 

Finally, the Elastic Net Regression model was created to experiment with different combinations of the Ridge and Lasso methods and to better handle highly correlated variables. When dealing with correlation, the Lasso regression model picks one variable and eliminates all the others, and the Ridge regression model tends to shrink all the correlated variables. The Elastic Net Regression can combine all the correlated variables and decide whether to shrink them or leave them out. Once again, Cross-validation was used to select the optimal alpha and lambda values. Alpha controls the mix of the Lasso and Ridge regression.



It can be observed that, despite testing different models, the RMSE value and R-squared values are indistinguishable, thus showing that the models' predictive accuracy is quite similar. However, given the above discussions about the advantages and drawbacks of each of the models, the Elastic Net Regression model would be an ideal model. This model heavily relied on Lasso regression, with the final alpha value being close to 1, and, therefore, includes the benefits of having a sparse model and feature selection that were discussed above. Compared to the OLS model, it is more reliable as the violations of the OLS model reduces its reliability.

However, heteroscedasticity error is present within the data itself that reduces the ability for the Elastic Net regression model to be improved further. Because of changing socioeconomic climates, the variability of the errors is not consistent across different variables. Therefore, the data was split into six different time segments, within which homoscedasticity holds. The Elastic Net Regression was then trained again for each segment separately.



Based on the results above, it can be seen that the RMSE values are significantly lower and the R-squared values are higher compared to when the model was trained on all the data together. It can be concluded that an Elastic Net regression model that is trained on separated time periods during which homoscedasticity is maintained, would be an ideal model to predict stock market volatility.

## Discussion

The goal of this paper was to design a model that is able to predict stock market volatility given social, political, and economic conditions. An OLS Regression model, a Lasso regression model, a Ridge regression model, and an Elastic Net regression model were trained on the available data. It was concluded that the Elastic Net regression model is the optimal model. While the predicted results were similar to the other models, it offers more advantages such as lower overfitting, higher sparsity, improved generalization, all the while maintaining model interpretability. Training the model on separate time segments also introduced homoscedasticity, which improved the results further.

The predictors with the largest coefficients that were statistically significant for the OLS regression model, and the largest lambda values in the Lasso, Ridge, and Elastic Net regression models, were inflation, interest rates, trade, business investment, and sentiment, regulation, and labor disputes. When examining the volatility of the stock market over the decades, the models unanimously agree that factors directly impacting business performance are the most important for stock market performance. It seems that social factors are either left out of the models or have a minor impact on volatility.

However, studying different time segments creates a different picture. The predictors chosen to have the biggest impact on the models closely align with historic events during those times. Between January 1990 and June 1998, the dot-com bubble was expanding, and the economy was in a period of growth. The factors most important in the model for this segment include real estate, financial and labor regulations, trade, healthcare policies, and electrical and political governance. All of these factors support a healthy labor force and financial condition. This was followed by the recession during July 1998 to March 2003, when the dot-com bubble burst. The model for this time segment highlighted labor disputes, fiscal policy, and food and drug policy, reflecting how people were struggling with their jobs, leading to concerns about affording basic necessities, and the need for governments to enact policies to improve the situation.

Another period of growth occurred between April 2003 and December 2007. The model for this time segment chose predictors like real estate, trade, intellectual property, taxes, entitlement and welfare programs, and elections and political governance. This shows how this was a time when there was an emphasis on increased support for a healthy labor force, research and development, and good business practices and entrepreneurship. The depression period caused by the housing crisis between December 2007 and June 2009 came next, and the model for this period selected political uncertainty, macroeconomic news and outlook, the labor market, the financial crisis, exchange rates, government spending, and multiple policies and regulations as important factors. This was a time when the labor force was highly uncertain about their future, the system that managed their finances and livelihood, and the government needed to enact numerous changes to handle the situation, with the importance of the exchange rate displaying the impact the stock market crash had on global trade.

From October 2009 to December 2019, there was a period of economic growth and increased human rights movements coupled with dividing government decisions, such as the 2016 American election and Brexit, increased mass shootings, and threats of global epidemics and pandemics, such as swine flu and COVID-19. This was reflected in the model, which prioritized infectious disease, intellectual property, immigration, elections, and political governance as important factors. Finally, during the period of the COVID-19 recession between January 2020 and December 2022, the model selected political uncertainty, consumer spending and sentiment, labor disputes, national security, housing and land management, transportation, infrastructure, and public utilities among the factors that most impacted the volatility of the stock market.

All of this clearly reflects the importance of the social and political situation on top of economic and business conditions on the impact on the stock market. Apart from business decisions and factors that impact trading practices, the health and well-being of the labor force, the political situation of an economy, consumer sentiment, and government support are all factors that impact economic conditions and, in turn, the volatility of the stock market. Factors highlighted by the general models, such as trade and interest rates, appeared among the models that focused on certain time segments as well, showing that they are factors that directly impact market conditions regardless of social and political conditions. It is interesting to note that intellectual property was missing for the model trained on the data from the time of the dot-com bubble, and real estate was missing for the model for the time segment during the housing crisis. While this may initially seem wrong, it makes sense as these are constant situations during the respective periods. Therefore, while it changed the volatility of the market compared to time segments before and after, the changes in the stock market volatility during the time were impacted by other factors.

A major drawback of all the models that were trained is that they all assume linearity between the predictors and VIX. This might not be the case when dealing with stock market volatility and the socioeconomic factors that impact it. Moving forward, it would be beneficial to use models such as neural networks and decision tree models to study the relation between these predictors and VIX as they would be able to better capture their relationship as they do not rely on linearity.

## Appendix

```r
library(tidyverse)
library(ggplot2)
library(e1071)
install.packages("caret")
library(caret)
install.packages("glmnet")
library(glmnet)
install.packages("ISLR")
library(ISLR)
library(readxl)
dataset <- read_excel("EMV_VIX_Data.xlsx")
y <- dataset$VIX
x <- model.matrix(VIX~.-Date, dataset)[,-1]
```

### OLS Regression

```r
ols <- lm(y ~ x)
predicted <- predict(ols)
residuals <- y - predicted
RMSE <- sqrt(mean(residuals^2))
RMSE
summary(ols)
```

```r
plot(ols$fit, ols$res, xlab = "Fitted", ylab = "Residuals", main = "Residuals vs fitted values")
```

### Lasso Regression

```r
set.seed(1)
lasso.cv = cv.glmnet(x, y, type.measure="mse", alpha=1, family="gaussian", nlamda=200, nfolds=10, standardize=TRUE)
lasso.cv$lambda.min
lasso.cv$lambda.1se
```

```r
round(cbind(coef(lasso.cv, s = 'lambda.min'), coef(lasso.cv, s = 'lambda.1se')), digits = 3)
```

```r
plot(lasso.cv)
abline(h = lasso.cv$cvup[lasso.cv$index[1]], lty=4)
```

```r
lasso = glmnet(x,y,alpha=1)
plot(lasso, xvar = 'lambda')
abline(v=log(lasso.cv$lambda.min), lty=3)
abline(v=log(lasso.cv$lambda.1se), lty=3)
```

```r
predictions_lasso <- glmnet(x,y,alpha=1, 
                             lambda = lasso.cv$lambda.min) %>% predict(x) %>% as.vector()

RMSE = RMSE(predictions_lasso, y)
RMSE

R_squared <- R2(predictions_lasso, y)
print(paste("R-squared:", R_squared))
```

### Ridge Regression

```r
set.seed(1)
ridge.cv = cv.glmnet(x, y, type.measure="mse", alpha=0, family="gaussian", nlamda=200, nfolds=10, standardize=TRUE)
ridge.cv$lambda.min
ridge.cv$lambda.1se
```

```r
round(cbind(coef(ridge.cv, s = 'lambda.min'), coef(ridge.cv, s = 'lambda.1se')), digits = 3)
```

```r
plot(ridge.cv)
abline(h = ridge.cv$cvup[ridge.cv$index[1]], lty=4)
```

```r
ridge = glmnet(x,y,alpha=0)
plot(ridge, xvar = 'lambda')
abline(v=log(ridge.cv$lambda.min), lty=3)
abline(v=log(ridge.cv$lambda.1se), lty=3)
```

```r
predictions_ridge <- glmnet(x,y,alpha=0, 
                             lambda = ridge.cv$lambda.min) %>% predict(x) %>% as.vector()

RMSE = RMSE(predictions_ridge, y)
RMSE

R_squared <- R2(predictions_ridge, y)
print(paste("R-squared:", R_squared))
```

### Elastic Net Regression

```r
k <- 10
set.seed(1)
train_control <- trainControl(method="cv", number = k)

elastic_model <- train(y ~ .,
                      data = cbind(x,y),
                      method = "glmnet",
                      preProcess = c("center", "scale"),
                      tuneLength = 10,
                      trControl = train_control)
print(elastic_model$finalModel$tuneValue)
```

```r
predictions_enet <- glmnet(x,y,alpha=elastic_model$finalModel$tuneValue$alpha,
                          lambda = elastic_model$finalModel$tuneValue$lambda) %>% predict(x) %>% as.vector()
RMSE_enet = RMSE(predictions_enet, y)
RMSE_enet

R_squared <- R2(predictions_enet, y)
print(paste("R-squared:", R_squared))
```

### 1990-01 - 1998-06

```r
if (!require(zoo)) {
  install.packages("zoo")
  library(zoo)
}
dataset <- read_excel("EMV_VIX_Data.xlsx")
dataset$Date <- as.yearmon(dataset$Date)
start_date <- as.yearmon("1990-01")
end_date <- as.yearmon("1998-06")
subset_data <- dataset[dataset$Date >= start_date & dataset$Date <= end_date, ]
y <- subset_data$VIX
x <- model.matrix(VIX ~ . - Date, data = subset_data)[, -1]
k <- 10
set.seed(1)
train_control <- trainControl(method="cv", number = k)

elastic_model <- train(y ~ .,
                      data = cbind(x,y),
                      method = "glmnet",
                      preProcess = c("center", "scale"),
                      tuneLength = 10,
                      trControl = train_control)
print(elastic_model$finalModel$tuneValue)

predictions_enet <- glmnet(x,y,alpha=elastic_model$finalModel$tuneValue$alpha,
                          lambda = elastic_model$finalModel$tuneValue$lambda) %>% predict(x) %>% as.vector()
RMSE_enet = RMSE(predictions_enet, y)
RMSE_enet
R_squared <- R2(predictions_enet, y)
print(paste("R-squared:", R_squared))

print("Variable names and lambda values:")
print(coef(elastic_model$finalModel, s = elastic_model$finalModel$tuneValue$lambda))
```
